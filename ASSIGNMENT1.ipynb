```python
import matplotlib.pyplot as plt

car_brand = ["Maruti", "Hyundai", "Honda", "Toyota"]
count = [25, 30, 40, 15]

plt.pie(count, labels=car_brand, autopct='%1.1f%%', startangle=90)
plt.title("Car Brand Distribution")
plt.axis('equal')  
plt.show()
```
